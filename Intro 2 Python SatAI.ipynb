## list comprehension

```python
x = [1,2,3,4]
```

```python
#Para calcular el cuadrado de cada elemento:

out = []
for item in x:
    out.append(item**2)

print(out)
```

```python
#Con list comprehension podemos resumirlo en una línea

[item**2 for item in x]


```

## map and filter
"map" aplica una función a cada elemento de una sequencia o iterable

*map(function, iterables)*


```python
seq = [1,2,3,4,5]
```

```python
map(times2,seq)
```

```python
#Lo convertemos a lista para ver el resultado

list(map(times2,seq))
```

```python
list(map(lambda var: var*2,seq))
```

"filter" entrega los valores para los cuales se cumple la condición



```python
filter(lambda item: item%2 == 0,seq)
```

```python
list(filter(lambda item: item%2 == 0,seq))

```
