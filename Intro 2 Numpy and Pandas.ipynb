## Dataframes
Dataframes son una de las herramientas más importantes para analizar data.
Se pueden imaginar como una tabla con índices y columnas

```python
A = pd.DataFrame(data = np.random.randn(3,3),
             index = ['a', 'b', 'c'],
             columns = ['col_a', 'col_b', 'col_c'])
A
```

```python
# Podemos acceder a los elementos de un dataframe
# con el índice y columna adecuado y con "loc"
A.loc['b', 'col_a']
```

```python
# También podemos usar la posición junto con "iloc"
# con "iloc"

A.iloc[1,0]
```

```python
#para obtener una columna
A['col_a']  #esto entrega una serie
```

```python
#para obtener más de una columna, se usan dos parentesis
A[['col_a', 'col_b']] #esto entrega un dataframe
```

Algo bastante útil de pandas es poder trabajar con condiciones. Por ejemplo:


```python
#Todos los elementos que cumplen la condición
A<0.5
```

```python
#Podemos recuperar las filas donde se cumple la condición
A.loc[A['col_b']<0.5, :]
```

```python
# Podemos combinar condiciones
A.loc[ (A['col_a']<0.3) & (A['col_a']>0), :]
```

```python
# y crear nuevos dataframes
B = A.loc[ (A['col_b']<0.5) & (A['col_a']<0.5)] * 0.5
B
```

```python
#Asignar nuevas columnas
A['almuerzo'] = ['gato','perro','hamster']
A
```

```python
#y definir un nuevo index
A.set_index('almuerzo')
```

```python
# NOTA: el cambio de índice (y algunas otras funciones) no cambian el dataframe,
# a menos que uno explícitamente lo requiera
A.set_index('almuerzo', inplace = True)
A
```

### Algunos métodos útiles de pandas
Una de las ventajas de pandas es que nos permite hacer cálculos de la data con gran facilidad.
Creemos un dataframe para ver algunos ejemplos:

```python
df = pd.DataFrame(data = [
['elephant',1001,500],
['elephant',1002,600],
['elephant',1003,550],
['tiger',1004,300],
['tiger',1005,320],
['tiger',1006,330],
['tiger',1007,290],
['tiger',1008,310],
['zebra',1009,200],
['zebra',1010,220],
['zebra',1011,240],
['zebra',1012,230],
['zebra',1013,220],
['zebra',1014,100],
['zebra',1015,80],
['lion',1016,420],
['lion',1017,600],
['lion',1018,500],
['lion',1019,390],
['kangaroo',1020,410],
['kangaroo',1021,430],
['kangaroo',1022,410]],
columns = ['animal','uniq_id','water_need'])

df
```

```python
# Para ver algunas filas

df.head(n=4)
```

```python
#Para información del dataframe

df.info()
```

```python
# Para las estadísticas de las columnas numéricas, podemos usar "describe"
df.describe()
```

```python
#O podemos hacer el cálculo por columna
df['water_need'].mean()
```

Hay métodos útiles para obtener información de las variables categóricas (string)


```python
# Elementos únicos
df['animal'].unique()
```

```python
# Número de elementos únicos
df['animal'].nunique()
```

```python
# Número de elementos por tipo
df['animal'].value_counts()
#df['animal'].value_counts(normalize = True)
```

```python
# Podemos agrupar por tipo de animal, y hacer cálculos
# Por ejemplo, la cantidad de agua necesaria promedio por tipo de animal

df.groupby('animal')['water_need'].mean()
```

```python
#Con pandas incluso podemos hacer gráficos
data = pd.DataFrame(data = 5*np.sin( np.arange(0,20) ) + np.random.randn(20),
             index = np.arange(0,20),
             columns = ['señal con ruido'])

data.plot()
```
