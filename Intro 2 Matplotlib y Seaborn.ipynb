```python
#Importamos matplotlib de esta manera

import matplotlib.pyplot as plt
```

### Ejemplos

Empecemos con algunos ejemplos simples

```python
import numpy as np
x = np.linspace(0, 5, 11) #eje X
y = x ** 2  #eje Y
```

```python
plt.plot(x, y, 'r') # 'r' para el color rojo (red)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Título de mi gráfico')
```

```python
#Podemos ajustar el gráfico si deseamos

plt.figure(figsize=(4,4))   #tamaño del gráfico
plt.scatter(x, y, color = 'r')      # scatterplot en vez de línae
plt.xlabel('Eje X', fontsize = 16)  #tamaño de la letra
plt.ylabel('Eje Y', fontsize = 16)
plt.title('Título de mi gráfico')
```

Podemos graficar en múltiples ventanas

```python
# plt.subplot(número de filas, número de columns, número del plot)
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-');
```

```python
#Podemos cambiar el estilo de las líneas
plt.figure()
plt.plot(x, x**2, 'r--', label = 'Cuadrado')    #label define la leyenda
plt.plot(x, x**3, 'b-.', label = 'Cubo')
plt.legend()
```

```python
#Gráfico de barras
data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

plt.bar(names, values)
```

```python
#Boxplot
import pandas as pd
data2 = pd.DataFrame(np.random.randn(100,3),
                     columns = ['A', 'B', 'C'])
data2
```

```python
#Podemos graficar con pandas

data2.boxplot()
```

```python
#O graficamos con matplotlib

plt.figure()
plt.boxplot(data2)
plt.show()
```

# Seaborn

Seaborn está basado en matplotlib, con la ventaja de que se enfoca
en gráficos especiales para análisis estadístico con un buen diseño

Por ejemplo, el mismo gráfico de boxplot, ahora en seaborn:

```python
!pip install seaborn
```

```python
import seaborn as sns

sns.boxplot(data = data2)
```

```python
#Veamos un ejemplo que viene con seaborn

sns.set_theme(style="ticks", palette="pastel")

# Load the example tips dataset
tips = sns.load_dataset("tips")
tips
```

```python
# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="day", y="total_bill",
            hue="smoker", palette=["m", "g"],
            data=tips)
#sns.despine(offset=10, trim=True)
```

```python
#Veamos las propinas
sns.boxplot(x="day", y="tip",
            hue="smoker", palette=["m", "g"],
            data=tips)
```

```python
#Podemos analizar la data en más detalle, por ejemplo, la distribución de la cuenta

sns.distplot(tips['total_bill'])
```

```python
# O comparar la distribución de dos variables
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#Si deseas, puedes ver el joint plot de todas las variables del dataframe
#Cuidado si son muchas variables, puede generar demasiados gráficos
sns.pairplot(tips)
```

```python
# Y también podemos especificar un hue
sns.pairplot(tips, hue = 'smoker')
```
