```python
import matplotlib.pyplot as plt
import seaborn as sns

titanic = sns.load_dataset('titanic')
titanic.head()

```

```python
#Distribución de edades

titanic['age'].hist()
plt.ylabel('Número de personas')
plt.xlabel('Edad')
plt.show()
```

```python
# Gráfico de barras de número de personas por género
titanic['sex'].value_counts().plot.bar()
plt.ylabel('número de personas')
plt.show()
```

```python
# Boxplots de edad por clase

sns.boxplot(data = titanic, x = 'class', y = 'age')
plt.show()
```

```python
# Boxplots de edad por clase con hue de sobreviviente

sns.boxplot(data = titanic, x = 'class', y = 'age', hue = 'survived')
plt.show()
```

```python
# Boxplots de fare por clase con hue de sobreviviente
# Nota: puede usar plt.ylim(lim_inferior,lim_superior) para
# definir los límites del eje Y

sns.boxplot(data = titanic, x = 'class', y = 'fare', hue = 'survived')
plt.ylim(0,200)
plt.show()
```

```python
# Jointplot de edad vs fare

sns.jointplot(data = titanic, x = 'age', y = 'fare')
plt.show()
```
