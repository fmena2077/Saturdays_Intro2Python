*Cree una función que, dado un email entregue el domain*

    user@domain.com
    
*Por ejemplo, al entregar "user@domain.com", la función entrega: domain.com*

```python
def domainGet(param):
  """
  param es el string input 
  """
  param_split = param.split('@')

  #El 2do elemento es el domain

  return param_split[-1]
```

```python
domainGet('user@domain.com')
```

*Cree una función que entrega True si la palabra 'dog' aparece en el input string. Tome en cuenta la capitalización del input string.*

```python
def findDog(frase):
  frase = frase.lower()   #bajar a minuscula
  frase = frase.split()   #separar por palabra
  if 'dog' in frase:
    return True
  else:
    return False
```

```python
findDog('Is there a dog here?')
```

*Cree una función que cuente el número de veces que aparece la palabra 'dog' en el string.*

```python
def countDog(frase):
  frase = frase.lower()   #bajar a minuscula
  frase = frase.split()   #separar por palabra
  contador = 0
  for palabra in frase:
    if palabra == 'dog':
      contador = contador +1

  return contador
```

```python
countDog('This dog runs faster than the other dog dude!')
```

### Problema Final
*Usted va conduciendo demasiado rápido y un oficial de policía lo detiene. Escriba una función
   para devolver uno de los 3 resultados posibles: "No Ticket", "Small Ticket" o "Big Ticket".
   Si su velocidad es de 60 o menos, el resultado es "No Ticket". Si la velocidad está entre 61
   y 80 inclusive, el resultado es "Small Ticket". Si la velocidad es 81 o más, el resultado es
   "Big Ticket". A menos que sea su cumpleaños! en
   su cumpleaños su velocidad puede ser 5 más alta en total
   casos (is_birthday = boolean)
  .*

```python
def caught_speeding(speed, is_birthday):
    
    #Primero veamos la condición del cumpleaños, que la velocidad se perdona en 5 unidades
    if is_birthday:
      speed = speed - 5

    #Ahora ponemos las condiciones:
    if speed <= 60: 
      return 'No Ticket'
    elif (speed >= 61) & (speed <= 80): 
      return 'Small Ticket'
    elif speed >= 81: 
      return 'Big Ticket'
    else:
      return 'Error, revisar input'
```

```python
caught_speeding(81,True)
```

```python
caught_speeding(81,False)
```
